# Account Transactions with Plaid

In this section, you will use the Plaid Python SDK to connect to the Developer Sandbox account and grab a list of transactions. You will need to complete the following steps:


1. Use the access token to fetch the transactions for the last 90 days
2. Print the categories for each transaction type
3. Create a new DataFrame using the following fields from the JSON transaction data: `date, name, amount, category`. (For categories with more than one label, just use the first category label in the list)
4. Convert the data types to the appropriate types (i.e. datetimeindex for the date and float for the amount)

### 1. Fetch the Transactions for the last 90 days

```python
# Get transactions for institution for last 90 days

start_date = '{:%Y-%m-%d}'.format(datetime.datetime.now() + datetime.timedelta(-90))
end_date = '{:%Y-%m-%d}'.format(datetime.datetime.now())

# Get transactions for date range

transaction_response = client.Transactions.get(access_token,start_date,end_date)

# Print JSON output

print(json.dumps(transaction_response['transactions'],indent=4, sort_keys=True))
```

### 2. Print the categories for each transaction

```python
# loop through transaction response and print off categories
count=0
for transactions in transaction_response['transactions']:
    print(json.dumps(transactions['category'], indent=4, sort_keys=True))
    count+=1
print(count)
```

### 3. Create a new DataFrame using the following fields from the JSON transaction data: date, name, amount, category. 

(For categories with more than one label, just use the first category label in the list)

```python


#loop through transactions and put data in lists, 

categories = []
dates = []
names = []
amounts = []

for transactions in transaction_response['transactions']:
    categories.append(json.dumps(transactions['category'][0]))
    dates.append(json.dumps(transactions['date'] ))
    names.append(json.dumps(transactions['name']))
    amounts.append(json.dumps(transactions['amount']))
    
#create new dataframe using the lists
transaction_df = pd.DataFrame([categories, dates, names, amounts])

#transpose dataframe to get get rows and columns correct
transaction_df = transaction_df.T

#rename columns and print head
transaction_df.columns = ['category','date', 'name', 'amount' ]
transaction_df.head(50)
    
```

### 4. Convert the data types to the appropriate types 

(i.e. datetimeindex for the date and float for the amount)

```python

#print off data types

print(transaction_df.dtypes)


#change date from string to datetime.  Why was my date formatted with two quotes? This cost me hours!

transaction_df['date'] = pd.to_datetime(transaction_df['date'], format= '"%Y-%m-%d"')


#transaction_df.set_index('date', inplace=True)

#change amount to float 

transaction_df['amount']= transaction_df['amount'].astype('float')

print(transaction_df.dtypes)

print(transaction_df.head(50))



```

# Income Analysis with Plaid

In this section, you will use the Plaid Sandbox to complete the following:
1. Determine the previous year's gross income and print the results
2. Determine the current monthly income and print the results
3. Determine the projected yearly income and print the results

```python

# Get income for date range

income_response = client.Income.get(access_token)


# Print last year's income from json response

last_year_income = (json.dumps(income_response['income']["last_year_income"],indent=4, sort_keys=True))
print(f"Last year's income was {last_year_income}")

#print current monthly income

monthly_income = (json.dumps(income_response['income']["income_streams"][0]["monthly_income"],indent=4, sort_keys=True))
print(f'Current monthly is {monthly_income}')

projected_yearly_income = (json.dumps(income_response['income']["projected_yearly_income"],indent=4, sort_keys=True))
print(f'The projected yearly income is {projected_yearly_income}')
```
